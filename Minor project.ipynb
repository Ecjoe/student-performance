```python
#Statistical analysis of student performance
student.gender.value_counts().plot(kind='bar')
plt.show()
```

```python
#Math score of students either male or female
plt.figure(figsize=(20,10))
sns.countplot(x='math_score',hue='gender',data=student)
plt.title('Math score',fontsize=20)
plt.show()
```

```python
#Reading score of students either male or female
plt.figure(figsize=(20,10))
sns.countplot(x='reading_score',hue='gender',data=student)
plt.title('Reading score',fontsize=20)
plt.show()
```

```python
#Writing score of students either male or female
plt.figure(figsize=(20,10))
sns.countplot(x='writing_score',hue='gender',data=student)
plt.title('Writing score',fontsize=20)
plt.show()
```

```python
#Showing highest performances of male or female students in math,reading or writing
plt.title('Math score',fontsize=20)
sns.barplot(y='math_score',x='gender',data=student,palette='bright')
plt.show()
plt.title('Reading score',fontsize=20)
sns.barplot(x='gender',y='reading_score',data=student,palette='bright')
plt.show()
plt.title('Writing score',fontsize=20)
sns.barplot(x='gender',y='writing_score',data=student,palette='bright')
plt.show()
```

Above plot shows that male students perform better at math whereas females at writing and reading.

```python
#Showing ethnicity category of students
plt.title('Race of students',fontsize=20)
sns.countplot(y='race',data=student)
plt.show()
```

Above plot shows that group C ethnic groups have maximum number of students. 

```python
#Showing parental education category of students
plt.title('Parental education of students',fontsize=20)
sns.countplot(y='parental_education',data=student)
plt.show()
```

Above plot shows that maximum number of students parental education is from some colleges. 

```python
#Showing lunch category of students
plt.title('Lunch of students',fontsize=20)
sns.countplot(x='lunch',data=student)
plt.show()
```

Maximun number of students eat standard lunch.

```python
#Showing test preparation of students
plt.title('Test preparation of students',fontsize=20)
sns.countplot(x='test_preparation',data=student)
plt.show()
```

Maximun number of students have taken no preparation course.

```python
#Showing which mode of test preparation affect the students performance either of male or female
plt.title("Effect of mode of test preparation on math score",fontsize=20)
sns.boxplot(x='test_preparation',y='math_score',hue='gender',data=student,palette='bright')
plt.show()
plt.title("Effect of mode of test preparation on reading score",fontsize=20)
sns.boxplot(x='test_preparation',y='reading_score',data=student,hue='gender',palette='bright')
plt.show()
plt.title("Effect of mode of test preparation on writing score",fontsize=20)
sns.boxplot(x='test_preparation',y='writing_score',data=student,hue='gender',palette='bright')
plt.show()
```

Above plot shows that males are better only at math instead females are best at reading and writing with or without test preparation course.

```python
#Shwoing which ethnic groups have the best perfornmance indicating gender 
plt.title("Effect of mode of race on math score",fontsize=20)
sns.boxplot(x='race',y='math_score',data=student,palette='bright',hue='gender')
plt.show()
plt.title("Effect of mode of race on reading score",fontsize=20)
sns.boxplot(x='race',y='reading_score',data=student,palette='bright',hue='gender')
plt.show()
plt.title("Effect of mode of race on writing score",fontsize=20)
sns.boxplot(x='race',y='writing_score',data=student,palette='bright',hue='gender')
plt.show()
```

Whether it is math,reading or writing,group E race have excellent skills and are best at them among which males are best at math and females at reading and writing.

```python
#Showing how lunch rates affect the student
plt.title("Effect of lunch on math score",fontsize=20)
sns.boxplot(x='lunch',y='math_score',data=student)
plt.show()
plt.title("Effect of lunch on reading score",fontsize=20)
sns.boxplot(x='lunch',y='reading_score',data=student)
plt.show()
plt.title("Effect of lunch on writing score",fontsize=20)
sns.boxplot(x='lunch',y='writing_score',data=student)
plt.show()
```

Maximum students prefer standard lunch which helps in gaining good scores at all subjects. 

```python
#Showing how student's parent background affects the performance
plt.title("Effect of parental education on math score",fontsize=20)
sns.boxplot(x='parental_education',y='math_score',data=student,hue='gender')
plt.xticks(fontsize=10,rotation=90)
plt.show()
```

The plot shows that parents,having master degres,their child,both females and males are good at maths.

```python
#Showing how student's parent background affects the performance
plt.title("Effect of parental education on writing score",fontsize=20)
sns.boxplot(x='parental_education',y='writing_score',data=student,hue='gender')
plt.xticks(fontsize=10,rotation=90)
plt.show()
```

The plot shows that parents,having master degres,their child,both females and males are good at writing.

```python
#Showing how student's parent background affects the performance
plt.title("Effect of parental education on reading score",fontsize=20)
sns.boxplot(x='parental_education',y='reading_score',data=student,hue='gender')
plt.xticks(fontsize=10,rotation=90)
plt.show()
```

The plot shows that parents,having master degres,their child,both females and males are good at reading.
